```python
from pathlib import Path
import numpy as np, pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display

import statsmodels.formula.api as smf

# ---------------------------------------------------------------------
# CONFIG
# ---------------------------------------------------------------------
IN_PARQUET = "cdr_keep_after_attn.parquet"
OUT_DIR = Path("analysis_out")
OUT_DIR.mkdir(exist_ok=True)

RQ1_ENV = [f"responsibility attri_{i}" for i in range(1, 9)]
RQ1_SOC = [f"resp_attribut_societ_{i}" for i in range(1, 9)]
RQ1_GOV = [f"resp_attribut_regula_{i}" for i in range(1, 9)]

ACTOR_LABELS = {
    1: "Tech providers (developers)",
    2: "Tech users (adopting firms)",
    3: "Employees",
    4: "Nat. govt & regulators",
    5: "EU govt & regulators",
    6: "Consumers",
    7: "NGOs / civil society",
    8: "News media & journalists",
}
ACTOR_ORDER = list(ACTOR_LABELS.values())

# ---------------------------------------------------------------------
# HELPERS (no jinja2 needed)
# ---------------------------------------------------------------------
def df_to_markdown(df, float_cols=None, digits=2):
    """Render a DataFrame as a Markdown table without needing jinja2/tabulate."""
    def esc(x):
        if pd.isna(x): return ""
        s = str(x).replace("|", "\\|").replace("\n", "<br>")
        return s
    df2 = df.copy()
    if float_cols:
        for c in float_cols:
            if c in df2.columns:
                df2[c] = df2[c].map(lambda v: (f"{v:.{digits}f}" if pd.notna(v) else ""))
    headers = [esc(c) for c in df2.columns]
    lines = []
    lines.append("| " + " | ".join(headers) + " |")
    lines.append("| " + " | ".join([":--" for _ in headers]) + " |")
    for _, row in df2.iterrows():
        lines.append("| " + " | ".join(esc(v) for v in row.values) + " |")
    return "\n".join(lines)

def show_table(df, float_cols=None, digits=2):
    display(Markdown(df_to_markdown(df, float_cols=float_cols, digits=digits)))

def fmt_p(p):
    if pd.isna(p): return "NA"
    return "<.001" if p < .001 else f"{p:.3f}"

# ---------------------------------------------------------------------
# LOAD DATA
# ---------------------------------------------------------------------
df = pd.read_parquet(IN_PARQUET)
print(f"Loaded {len(df):,} rows from {IN_PARQUET}")

# choose respondent ID (or synthetic)
if "ResponseId" in df.columns:
    rid = df["ResponseId"].astype(str)
else:
    rid = pd.Series(np.arange(len(df)), dtype=str, name="synthetic_id")

# ---------------------------------------------------------------------
# RESHAPE LONG (ESG Ã— actor)
# ---------------------------------------------------------------------
def build_long(domain, cols):
    rows = []
    for i, col in enumerate(cols, start=1):
        if col in df.columns:
            s = pd.to_numeric(df[col], errors="coerce")
            tmp = pd.DataFrame({
                "resp_id": rid,
                "domain": domain,
                "actor_idx": i,
                "actor": ACTOR_LABELS[i],
                "value": s,
            })
            rows.append(tmp)
    return pd.concat(rows, ignore_index=True) if rows else pd.DataFrame()

long = pd.concat([
    build_long("Environmental", RQ1_ENV),
    build_long("Social",        RQ1_SOC),
    build_long("Governance",    RQ1_GOV)
], ignore_index=True)

# filter valid Likert 1â€“7
long = long[(long["value"] >= 1) & (long["value"] <= 7)]
long["domain"] = pd.Categorical(long["domain"], categories=["Environmental","Social","Governance"])
long["actor"] = pd.Categorical(long["actor"], categories=ACTOR_ORDER)

# ---------------------------------------------------------------------
# DESCRIPTIVES
# ---------------------------------------------------------------------
desc = (
    long.groupby(["domain","actor"], as_index=False)
        .agg(mean=("value","mean"), sd=("value","std"), n=("value","size"))
)
desc["se"] = desc["sd"] / np.sqrt(desc["n"])
tcrit = stats.t.ppf(0.975, (desc["n"]-1).clip(lower=1))
desc["ci_low"] = desc["mean"] - tcrit * desc["se"]
desc["ci_high"] = desc["mean"] + tcrit * desc["se"]
desc = desc.sort_values(["domain","mean"], ascending=[True, False])

desc_path = OUT_DIR / "rq1_attributions_descriptives.csv"
desc.to_csv(desc_path, index=False)

display(Markdown("### ðŸ“Š RQ1 â€” Descriptives per domein Ã— actor"))
show_table(desc, float_cols=["mean","sd","se","ci_low","ci_high"], digits=2)

# ---------------------------------------------------------------------
# MODEL (cluster-robust OLS)
# ---------------------------------------------------------------------
ols = smf.ols("value ~ C(domain) * C(actor)", data=long).fit(
    cov_type="cluster", cov_kwds={"groups": long["resp_id"]}
)
display(Markdown(
    "### ðŸ§ª Model: value ~ domain * actor (cluster-robust SEs)\n"
    f"- n rows: {len(long):,}; respondents: {long['resp_id'].nunique():,}<br>"
    f"- RÂ² = {ols.rsquared:.3f}"
))

# Planned contrasts: within each domain, actor vs Tech providers
def within_domain_contrasts(domain):
    sub = long[long["domain"] == domain]
    fit = smf.ols("value ~ C(actor)", data=sub).fit(cov_type="cluster", cov_kwds={"groups": sub["resp_id"]})
    rows = []
    ref = "Tech providers (developers)"
    for a in ACTOR_ORDER:
        if a == ref: continue
        cname = f"C(actor)[T.{a}]"
        diff = fit.params.get(cname, np.nan)
        se = fit.bse.get(cname, np.nan)
        tval = diff / se if (pd.notna(se) and se != 0) else np.nan
        pval = 2 * stats.t.sf(abs(tval), df=fit.df_resid) if pd.notna(tval) else np.nan
        rows.append({"domain": domain, "contrast": f"{a} âˆ’ {ref}", "diff": diff, "se": se, "t": tval, "p": pval})
    return pd.DataFrame(rows)

contrasts = pd.concat([within_domain_contrasts(d) for d in ["Environmental","Social","Governance"]], ignore_index=True)
contrasts["p_fmt"] = contrasts["p"].map(fmt_p)
contrasts_path = OUT_DIR / "rq1_attributions_contrasts.csv"
contrasts.to_csv(contrasts_path, index=False)

display(Markdown("#### Planned contrasts (actor vs tech providers)"))
show_table(
    contrasts[["domain","contrast","diff","se","t","p_fmt"]]
      .assign(diff=lambda x: x["diff"].round(2),
              se=lambda x: x["se"].round(2),
              t=lambda x: x["t"].round(2)),
    float_cols=["diff","se","t"], digits=2
)

# Companies (providers+users) vs others
long["is_company"] = long["actor"].isin(["Tech providers (developers)","Tech users (adopting firms)"]).astype(int)
rows = []
for dom in ["Environmental","Social","Governance"]:
    sub = long[long["domain"] == dom]
    fit = smf.ols("value ~ is_company", data=sub).fit(cov_type="cluster", cov_kwds={"groups": sub["resp_id"]})
    b = fit.params.get("is_company", np.nan); se = fit.bse.get("is_company", np.nan)
    tval = b / se if (pd.notna(se) and se != 0) else np.nan
    pval = 2 * stats.t.sf(abs(tval), df=fit.df_resid) if pd.notna(tval) else np.nan
    g1 = sub[sub["is_company"]==1]["value"]; g0 = sub[sub["is_company"]==0]["value"]
    sp = np.sqrt(((len(g1)-1)*g1.var(ddof=1) + (len(g0)-1)*g0.var(ddof=1)) / max(len(g1)+len(g0)-2, 1))
    d_eff = (g1.mean() - g0.mean()) / sp if sp>0 else np.nan
    rows.append({"domain": dom, "companies_minus_others": b, "se": se,
                 "t": tval, "p": pval, "cohens_d": d_eff,
                 "mean_companies": g1.mean(), "mean_others": g0.mean()})
companies = pd.DataFrame(rows)
companies_path = OUT_DIR / "rq1_companies_vs_others.csv"
companies.to_csv(companies_path, index=False)

display(Markdown("#### Companies (providers + users) vs others"))
show_table(
    companies.assign(
        companies_minus_others=lambda x: x["companies_minus_others"].round(2),
        se=lambda x: x["se"].round(2),
        t=lambda x: x["t"].round(2),
        p=lambda x: x["p"].map(fmt_p),
        cohens_d=lambda x: x["cohens_d"].round(2),
        mean_companies=lambda x: x["mean_companies"].round(2),
        mean_others=lambda x: x["mean_others"].round(2),
    ),
    float_cols=["companies_minus_others","se","t","cohens_d","mean_companies","mean_others"],
    digits=2
)

# ---------------------------------------------------------------------
# FIGURES
# ---------------------------------------------------------------------
for dom in ["Environmental","Social","Governance"]:
    sub = desc[desc["domain"]==dom]
    plt.figure(figsize=(9,5))
    xpos = np.arange(len(sub))
    plt.bar(xpos, sub["mean"])
    for x, low, high in zip(xpos, sub["ci_low"], sub["ci_high"]):
        plt.plot([x,x],[low,high], color="black")
        plt.plot([x-0.1,x+0.1],[low,low], color="black")
        plt.plot([x-0.1,x+0.1],[high,high], color="black")
    plt.xticks(xpos, sub["actor"], rotation=35, ha="right")
    plt.ylabel("Responsibility (1â€“7)")
    plt.ylim(1,7)
    plt.title(f"Responsibility by actor â€” {dom}")
    plt.tight_layout()
    plt.savefig(OUT_DIR / f"rq1_{dom.lower()}_bars.png", dpi=200)
    plt.show()

print(f"âœ… Done. Outputs written to: {OUT_DIR.resolve()}")
```
