# RQ1 â€“ Digital Responsibility Attribution


```
RQ1: How is digital responsibility attributed to companies compared to other key actors across the Environmental, Social, and Governance (ESG) domains?
```

*Purpose*: Compare mean responsibility attributions across actors (companies vs. others) and across domains (Environmental, Social, Governance).

````python
# -*- coding: utf-8 -*-
# ============================================================
# RQ1 â€” Digital responsibility attributions by actor and domain
# - Descriptives (means, SDs, CIs)
# - Cluster-robust OLS models
# - Planned contrasts (actor vs tech providers)
# - Companies (providers+users) vs Others: (M, SD, n) + tests + Holm
# - Figures saved to Research Drive
# ============================================================

import numpy as np, pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display
import io
import statsmodels.formula.api as smf
from statsmodels.stats.multitest import multipletests

# --- WebDAV helpers
from rd_io import (
    OUT_DIR, rd_join,
    rd_read_parquet_df, rd_write_csv_df, rd_upload_bytes
)

# ------------------------------------------------------------
# CONFIG (remote)
# ------------------------------------------------------------
IN_PARQUET = rd_join(OUT_DIR, "cdr_keep_after_attn.parquet")

RQ1_ENV = [f"responsibility attri_{i}" for i in range(1, 9)]
RQ1_SOC = [f"resp_attribut_societ_{i}" for i in range(1, 9)]
RQ1_GOV = [f"resp_attribut_regula_{i}" for i in range(1, 9)]

ACTOR_LABELS = {
    1: "Tech providers (developers)",
    2: "Tech users (adopting firms)",
    3: "Employees",
    4: "Nat. govt & regulators",
    5: "EU govt & regulators",
    6: "Consumers",
    7: "NGOs / civil society",
    8: "News media & journalists",
}
ACTOR_ORDER = list(ACTOR_LABELS.values())
DOMAINS = ["Environmental","Social","Governance"]

# ------------------------------------------------------------
# HELPERS
# ------------------------------------------------------------
def df_to_markdown(df, float_cols=None, digits=2):
    def esc(x):
        if pd.isna(x): return ""
        return str(x).replace("|","\\|").replace("\n","<br>")
    df2 = df.copy()
    if float_cols:
        for c in float_cols:
            if c in df2.columns:
                df2[c] = df2[c].map(lambda v: (f"{v:.{digits}f}" if pd.notna(v) else ""))
    headers = [esc(c) for c in df2.columns]
    lines = ["| " + " | ".join(headers) + " |",
             "| " + " | ".join([":--" for _ in headers]) + " |"]
    for _, row in df2.iterrows():
        lines.append("| " + " | ".join(esc(v) for v in row.values) + " |")
    return "\n".join(lines)

def show_table(df, float_cols=None, digits=2):
    display(Markdown(df_to_markdown(df, float_cols=float_cols, digits=digits)))

def fmt_p(p):
    if pd.isna(p): return "NA"
    return "<.001" if p < .001 else f"{p:.3f}"

def savefig_remote(fig, remote_path: str, fmt: str | None = None, dpi: int = 200):
    """Save a Matplotlib figure directly to Research Drive."""
    if fmt is None:
        fmt = remote_path.split(".")[-1].lower()
    buf = io.BytesIO()
    fig.savefig(buf, format=fmt, dpi=dpi, bbox_inches="tight")
    ct = "image/png" if fmt == "png" else "image/svg+xml" if fmt == "svg" else "application/octet-stream"
    rd_upload_bytes(remote_path, buf.getvalue(), ct)
    buf.close()

# ------------------------------------------------------------
# LOAD (remote)
# ------------------------------------------------------------
df = rd_read_parquet_df(IN_PARQUET)
print(f"Loaded {len(df):,} rows from {IN_PARQUET}")

# Respondent id (for clustering)
rid = df["ResponseId"].astype(str) if "ResponseId" in df.columns else pd.Series(np.arange(len(df)), dtype=str, name="synthetic_id")

# ------------------------------------------------------------
# RESHAPE LONG (ESG Ã— actor)
# ------------------------------------------------------------
def build_long(domain, cols):
    rows = []
    for i, col in enumerate(cols, start=1):
        if col in df.columns:
            s = pd.to_numeric(df[col], errors="coerce")
            rows.append(pd.DataFrame({
                "resp_id": rid,
                "domain": domain,
                "actor_idx": i,
                "actor": ACTOR_LABELS[i],
                "value": s,
            }))
    return pd.concat(rows, ignore_index=True) if rows else pd.DataFrame()

long = pd.concat([
    build_long("Environmental", RQ1_ENV),
    build_long("Social",        RQ1_SOC),
    build_long("Governance",    RQ1_GOV)
], ignore_index=True)

# filter valid Likert 1â€“7
long = long[(long["value"] >= 1) & (long["value"] <= 7)]
long["domain"] = pd.Categorical(long["domain"], categories=DOMAINS)
long["actor"] = pd.Categorical(long["actor"], categories=ACTOR_ORDER)

# ------------------------------------------------------------
# DESCRIPTIVES per domain Ã— actor
# ------------------------------------------------------------
desc = (
    long.groupby(["domain","actor"], as_index=False)
        .agg(mean=("value","mean"), sd=("value","std"), n=("value","size"))
)
desc["se"] = desc["sd"] / np.sqrt(desc["n"])
tcrit = stats.t.ppf(0.975, (desc["n"]-1).clip(lower=1))
desc["ci_low"] = desc["mean"] - tcrit * desc["se"]
desc["ci_high"] = desc["mean"] + tcrit * desc["se"]
desc = desc.sort_values(["domain","mean"], ascending=[True, False])

rd_write_csv_df(desc, rd_join(OUT_DIR, "rq1_attributions_descriptives.csv"), index=False)

display(Markdown("### ðŸ“Š RQ1 â€” Descriptives per domein Ã— actor"))
show_table(desc, float_cols=["mean","sd","se","ci_low","ci_high"], digits=2)

# ------------------------------------------------------------
# GLOBAL MODEL (cluster-robust OLS)
# ------------------------------------------------------------
ols = smf.ols("value ~ C(domain) * C(actor)", data=long).fit(
    cov_type="cluster", cov_kwds={"groups": long["resp_id"]}
)
display(Markdown(
    "### ðŸ§ª Model: value ~ domain * actor (cluster-robust SEs)\n"
    f"- n rows: {len(long):,}; respondents: {long['resp_id'].nunique():,}<br>"
    f"- RÂ² = {ols.rsquared:.3f}"
))

# ------------------------------------------------------------
# PLANNED CONTRASTS: within each domain, actor vs Tech providers
# ------------------------------------------------------------
def within_domain_contrasts(domain):
    sub = long[long["domain"] == domain]
    fit = smf.ols("value ~ C(actor)", data=sub).fit(cov_type="cluster", cov_kwds={"groups": sub["resp_id"]})
    rows = []
    ref = "Tech providers (developers)"
    for a in ACTOR_ORDER:
        if a == ref: continue
        cname = f"C(actor)[T.{a}]"
        diff = fit.params.get(cname, np.nan)
        se = fit.bse.get(cname, np.nan)
        tval = diff / se if (pd.notna(se) and se != 0) else np.nan
        pval = 2 * stats.t.sf(abs(tval), df=fit.df_resid) if pd.notna(tval) else np.nan
        rows.append({"domain": domain, "contrast": f"{a} âˆ’ {ref}", "diff": diff, "se": se, "t": tval, "p": pval})
    return pd.DataFrame(rows)

contrasts = pd.concat([within_domain_contrasts(d) for d in DOMAINS], ignore_index=True)
contrasts["p_fmt"] = contrasts["p"].map(fmt_p)
rd_write_csv_df(contrasts, rd_join(OUT_DIR, "rq1_attributions_contrasts.csv"), index=False)

display(Markdown("#### Planned contrasts (actor vs tech providers)"))
show_table(
    contrasts[["domain","contrast","diff","se","t","p_fmt"]]
      .assign(diff=lambda x: x["diff"].round(2),
              se=lambda x: x["se"].round(2),
              t=lambda x: x["t"].round(2)),
    float_cols=["diff","se","t"], digits=2
)

# ------------------------------------------------------------
# COMPANIES (providers+users) vs OTHERS â€” tests + (M, SD, n)
# ------------------------------------------------------------
long["is_company"] = long["actor"].isin([
    "Tech providers (developers)","Tech users (adopting firms)"
]).astype(int)

rows = []
for dom in DOMAINS:
    sub = long[long["domain"] == dom]
    fit = smf.ols("value ~ is_company", data=sub).fit(cov_type="cluster", cov_kwds={"groups": sub["resp_id"]})
    b = fit.params.get("is_company", np.nan)
    se = fit.bse.get("is_company", np.nan)
    tval = b / se if (pd.notna(se) and se != 0) else np.nan
    pval = 2 * stats.t.sf(abs(tval), df=fit.df_resid) if pd.notna(tval) else np.nan

    g1 = sub[sub["is_company"]==1]["value"]  # Companies rows
    g0 = sub[sub["is_company"]==0]["value"]  # Others rows

    # Cohen's d (pooled SD)
    sp = np.sqrt(((len(g1)-1)*g1.var(ddof=1) + (len(g0)-1)*g0.var(ddof=1)) / max(len(g1)+len(g0)-2, 1))
    d_eff = (g1.mean() - g0.mean()) / sp if sp>0 else np.nan

    rows.append({
        "domain": dom,
        "companies_minus_others": b,
        "se": se,
        "t": tval,
        "p": pval,
        "cohens_d": d_eff,
        "mean_companies": g1.mean(),
        "sd_companies": g1.std(ddof=1),
        "n_companies": int(g1.shape[0]),
        "mean_others": g0.mean(),
        "sd_others": g0.std(ddof=1),
        "n_others": int(g0.shape[0]),
    })

companies = pd.DataFrame(rows)

# Holm correction across the 3 domain tests
rej, p_holm, _, _ = multipletests(companies["p"].astype(float).values, method="holm")
companies["p_holm"] = p_holm
companies["sig_holm"] = rej

# Save & display
rd_write_csv_df(companies, rd_join(OUT_DIR, "rq1_companies_vs_others_M_SD_n.csv"), index=False)

display(Markdown("#### Companies (providers + users) vs Others â€” with (M, SD, n)"))
show_table(
    companies.assign(
        companies_minus_others=lambda x: x["companies_minus_others"].round(2),
        se=lambda x: x["se"].round(2),
        t=lambda x: x["t"].round(2),
        p=lambda x: x["p"].map(fmt_p),
        p_holm=lambda x: x["p_holm"].map(fmt_p),
        cohens_d=lambda x: x["cohens_d"].round(2),
        mean_companies=lambda x: x["mean_companies"].round(2),
        sd_companies=lambda x: x["sd_companies"].round(2),
        mean_others=lambda x: x["mean_others"].round(2),
        sd_others=lambda x: x["sd_others"].round(2),
    )[
        ["domain",
         "mean_companies","sd_companies","n_companies",
         "mean_others","sd_others","n_others",
         "companies_minus_others","se","t","p","p_holm","cohens_d"]
    ],
    float_cols=["mean_companies","sd_companies","mean_others","sd_others",
                "companies_minus_others","se","t","cohens_d"],
    digits=2
)


print("\nResults:")
for _, r in companies.iterrows():
    print(
        f"{r['domain']}: Companies (M={r['mean_companies']:.2f}, SD={r['sd_companies']:.2f}, n={int(r['n_companies'])}) "
        f"vs Others (M={r['mean_others']:.2f}, SD={r['sd_others']:.2f}, n={int(r['n_others'])}). "
        f"Cluster-robust OLS: Î”={r['companies_minus_others']:.2f} (SE={r['se']:.2f}), "
        f"t={r['t']:.2f}, p={fmt_p(r['p'])} (Holm={fmt_p(r['p_holm'])}), d={r['cohens_d']:.2f}."
    )

# ------------------------------------------------------------
# FIGURES (saved remotely)
# ------------------------------------------------------------
# A) Per-domain bars with 95% CI for every actor
for dom in DOMAINS:
    sub = desc[desc["domain"]==dom]
    fig = plt.figure(figsize=(9,5))
    ax = plt.gca()
    xpos = np.arange(len(sub))
    ax.bar(xpos, sub["mean"])
    for x, low, high in zip(xpos, sub["ci_low"], sub["ci_high"]):
        ax.plot([x,x],[low,high])
        ax.plot([x-0.1,x+0.1],[low,low])
        ax.plot([x-0.1,x+0.1],[high,high])
    ax.set_xticks(xpos, sub["actor"], rotation=35, ha="right")
    ax.set_ylabel("Responsibility (1â€“7)")
    ax.set_ylim(1,7)
    ax.set_title(f"Responsibility by actor â€” {dom}")
    fig.tight_layout()
    savefig_remote(fig, rd_join(OUT_DIR, f"rq1_{dom.lower()}_bars.png"))
    plt.show()

# B) Companies vs Others across domains â€” grouped means plot (optional)
#    Uses normal CIs around group means (descriptive), tests are reported above
viz_rows = []
for dom in DOMAINS:
    sub = long[long["domain"]==dom].copy()
    g1 = sub[sub["is_company"]==1]["value"]
    g0 = sub[sub["is_company"]==0]["value"]
    # Mean and 95% CI (descriptive)
    for label, s in [("Companies", g1), ("Others", g0)]:
        m = s.mean()
        se = s.std(ddof=1)/np.sqrt(s.shape[0])
        tcrit = stats.t.ppf(0.975, df=max(s.shape[0]-1,1))
        viz_rows.append({
            "domain": dom, "group": label, "mean": m,
            "ci_low": m - tcrit*se, "ci_high": m + tcrit*se
        })

viz = pd.DataFrame(viz_rows)

fig = plt.figure(figsize=(8,5))
ax = plt.gca()
x = np.arange(len(DOMAINS))
w = 0.35

m_comp = viz[viz["group"]=="Companies"].set_index("domain").loc[DOMAINS]
m_oth  = viz[viz["group"]=="Others"].set_index("domain").loc[DOMAINS]

ax.bar(x - w/2, m_comp["mean"].values, width=w, label="Companies")
ax.bar(x + w/2, m_oth["mean"].values,  width=w, label="Others")

# error bars
for xi, (lo, hi) in zip(x - w/2, zip(m_comp["ci_low"].values, m_comp["ci_high"].values)):
    ax.plot([xi, xi], [lo, hi])
    ax.plot([xi-0.08, xi+0.08], [lo, lo]); ax.plot([xi-0.08, xi+0.08], [hi, hi])
for xi, (lo, hi) in zip(x + w/2, zip(m_oth["ci_low"].values, m_oth["ci_high"].values)):
    ax.plot([xi, xi], [lo, hi])
    ax.plot([xi-0.08, xi+0.08], [lo, lo]); ax.plot([xi-0.08, xi+0.08], [hi, hi])

ax.set_xticks(x, DOMAINS)
ax.set_ylim(1,7)
ax.set_ylabel("Perceived responsibility (1â€“7)")
ax.set_title("Companies vs Others across ESG (means with 95% CI)")
ax.legend(frameon=False)
fig.tight_layout()
savefig_remote(fig, rd_join(OUT_DIR, "rq1_companies_vs_others_bars_all_domains.png"))
plt.show()

print(f"âœ… Done. Outputs written under remote OUT_DIR: {OUT_DIR}")
````

````python
# -*- coding: utf-8 -*-
# ============================================================
# RQ1 â€” Attribution of digital responsibility by actor & domain
# - Load & reshape to long
# - Descriptives (means, SDs, CIs) per domain Ã— actor
# - Cluster-robust OLS: Companies vs Others (with M, SD, n)
# - Paired t-tests per domain (Companies vs Others) + Holm + CIs + d_z
# - 2Ã—3 Repeated-measures ANOVA (Group Ã— Domain) + partial eta^2
# - APA-style lines (safe string formatting)
# ============================================================

import warnings
warnings.filterwarnings(
    "ignore",
    category=FutureWarning,
    message="The default of observed=False is deprecated.*"
)

import numpy as np, pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display
import io
import statsmodels.formula.api as smf
from statsmodels.stats.multitest import multipletests
from statsmodels.stats.anova import AnovaRM

# --- WebDAV helpers (provided in your environment)
from rd_io import (
    OUT_DIR, rd_join,
    rd_read_parquet_df, rd_write_csv_df, rd_upload_bytes
)

# ------------------------------------------------------------
# CONFIG
# ------------------------------------------------------------
IN_PARQUET = rd_join(OUT_DIR, "cdr_keep_after_attn.parquet")

RQ1_ENV = [f"responsibility attri_{i}" for i in range(1, 9)]
RQ1_SOC = [f"resp_attribut_societ_{i}" for i in range(1, 9)]
RQ1_GOV = [f"resp_attribut_regula_{i}" for i in range(1, 9)]

ACTOR_LABELS = {
    1: "Tech providers (developers)",
    2: "Tech users (adopting firms)",
    3: "Employees",
    4: "Nat. govt & regulators",
    5: "EU govt & regulators",
    6: "Consumers",
    7: "NGOs / civil society",
    8: "News media & journalists",
}
ACTOR_ORDER = list(ACTOR_LABELS.values())
DOMAINS = ["Environmental","Social","Governance"]
COMPANY_SET = {"Tech providers (developers)", "Tech users (adopting firms)"}

# ------------------------------------------------------------
# HELPERS
# ------------------------------------------------------------
def df_to_markdown(df, float_cols=None, digits=2):
    def esc(x):
        if pd.isna(x): return ""
        return str(x).replace("|","\\|").replace("\n","<br>")
    df2 = df.copy()
    if float_cols:
        for c in float_cols:
            if c in df2.columns:
                df2[c] = df2[c].map(lambda v: (f"{v:.{digits}f}" if pd.notna(v) else ""))
    headers = [esc(c) for c in df2.columns]
    lines = ["| " + " | ".join(headers) + " |",
             "| " + " | ".join([":--" for _ in headers]) + " |"]
    for _, row in df2.iterrows():
        lines.append("| " + " | ".join(esc(v) for v in row.values) + " |")
    return "\n".join(lines)

def show_table(df, float_cols=None, digits=2):
    display(Markdown(df_to_markdown(df, float_cols=float_cols, digits=digits)))

def fmt_p(p):
    if pd.isna(p): return "NA"
    return "<.001" if p < .001 else f"{p:.3f}"

def savefig_remote(fig, remote_path: str, fmt: str | None = None, dpi: int = 200):
    """Save a Matplotlib figure directly to Research Drive."""
    if fmt is None:
        fmt = remote_path.split(".")[-1].lower()
    buf = io.BytesIO()
    fig.savefig(buf, format=fmt, dpi=dpi, bbox_inches="tight")
    ct = "image/png" if fmt == "png" else "image/svg+xml" if fmt == "svg" else "application/octet-stream"
    rd_upload_bytes(remote_path, buf.getvalue(), ct)
    buf.close()

# ------------------------------------------------------------
# LOAD
# ------------------------------------------------------------
df = rd_read_parquet_df(IN_PARQUET)
print(f"Loaded {len(df):,} rows from {IN_PARQUET}")

# Respondent id (for clustering)
rid = df["ResponseId"].astype(str) if "ResponseId" in df.columns else pd.Series(np.arange(len(df)), dtype=str, name="synthetic_id")

# ------------------------------------------------------------
# RESHAPE LONG (ESG Ã— actor)
# ------------------------------------------------------------
def build_long(domain, cols):
    rows = []
    for i, col in enumerate(cols, start=1):
        if col in df.columns:
            s = pd.to_numeric(df[col], errors="coerce")
            rows.append(pd.DataFrame({
                "resp_id": rid,
                "domain": domain,
                "actor_idx": i,
                "actor": ACTOR_LABELS[i],
                "value": s,
            }))
    return pd.concat(rows, ignore_index=True) if rows else pd.DataFrame()

long = pd.concat([
    build_long("Environmental", RQ1_ENV),
    build_long("Social",        RQ1_SOC),
    build_long("Governance",    RQ1_GOV)
], ignore_index=True)

# Filter valid Likert 1â€“7 and set categories
long = long[(long["value"] >= 1) & (long["value"] <= 7)]
long["domain"] = pd.Categorical(long["domain"], categories=DOMAINS)
long["actor"] = pd.Categorical(long["actor"], categories=ACTOR_ORDER)

# ------------------------------------------------------------
# DESCRIPTIVES per domain Ã— actor
# ------------------------------------------------------------
desc = (
    long.groupby(["domain","actor"], as_index=False, observed=False)
        .agg(mean=("value","mean"), sd=("value","std"), n=("value","size"))
)
desc["se"] = desc["sd"] / np.sqrt(desc["n"])
tcrit_desc = stats.t.ppf(0.975, (desc["n"]-1).clip(lower=1))
desc["ci_low"] = desc["mean"] - tcrit_desc * desc["se"]
desc["ci_high"] = desc["mean"] + tcrit_desc * desc["se"]
desc = desc.sort_values(["domain","mean"], ascending=[True, False])

rd_write_csv_df(desc, rd_join(OUT_DIR, "rq1_attributions_descriptives.csv"), index=False)

display(Markdown("### ðŸ“Š RQ1 â€” Descriptives per domain Ã— actor"))
show_table(desc, float_cols=["mean","sd","se","ci_low","ci_high"], digits=2)

# ------------------------------------------------------------
# CLUSTER-ROBUST OLS: Companies (providers+users) vs Others
# ------------------------------------------------------------
long["is_company"] = long["actor"].isin(COMPANY_SET).astype(int)

rows_cr = []
for dom in DOMAINS:
    sub = long[long["domain"] == dom]
    fit = smf.ols("value ~ is_company", data=sub).fit(
        cov_type="cluster", cov_kwds={"groups": sub["resp_id"]}
    )
    b = fit.params.get("is_company", np.nan)
    se = fit.bse.get("is_company", np.nan)
    tval = b / se if (pd.notna(se) and se != 0) else np.nan
    pval = 2 * stats.t.sf(abs(tval), df=fit.df_resid) if pd.notna(tval) else np.nan

    g1 = sub[sub["is_company"]==1]["value"]  # Companies
    g0 = sub[sub["is_company"]==0]["value"]  # Others

    # Cohen's d (pooled SD)
    sp = np.sqrt(((len(g1)-1)*g1.var(ddof=1) + (len(g0)-1)*g0.var(ddof=1)) / max(len(g1)+len(g0)-2, 1))
    d_eff = (g1.mean() - g0.mean()) / sp if sp>0 else np.nan

    rows_cr.append({
        "domain": dom,
        "companies_minus_others": b,
        "se": se,
        "t": tval,
        "p": pval,
        "cohens_d": d_eff,
        "mean_companies": g1.mean(),
        "sd_companies": g1.std(ddof=1),
        "n_companies": int(g1.shape[0]),
        "mean_others": g0.mean(),
        "sd_others": g0.std(ddof=1),
        "n_others": int(g0.shape[0]),
    })

companies_cr = pd.DataFrame(rows_cr)

# Holm correction across the 3 domain tests
rej_cr, p_holm_cr, _, _ = multipletests(companies_cr["p"].astype(float).values, method="holm")
companies_cr["p_holm"] = p_holm_cr
companies_cr["sig_holm"] = rej_cr

rd_write_csv_df(companies_cr, rd_join(OUT_DIR, "rq1_companies_vs_others_M_SD_n_cluster_ols.csv"), index=False)

display(Markdown("#### Companies (providers + users) vs Others â€” cluster-robust OLS with (M, SD, n)"))
show_table(
    companies_cr.assign(
        companies_minus_others=lambda x: x["companies_minus_others"].round(2),
        se=lambda x: x["se"].round(2),
        t=lambda x: x["t"].round(2),
        p=lambda x: x["p"].map(fmt_p),
        p_holm=lambda x: x["p_holm"].map(fmt_p),
        cohens_d=lambda x: x["cohens_d"].round(2),
        mean_companies=lambda x: x["mean_companies"].round(2),
        sd_companies=lambda x: x["sd_companies"].round(2),
        mean_others=lambda x: x["mean_others"].round(2),
        sd_others=lambda x: x["sd_others"].round(2),
    )[
        ["domain",
         "mean_companies","sd_companies","n_companies",
         "mean_others","sd_others","n_others",
         "companies_minus_others","se","t","p","p_holm","cohens_d"]
    ],
    float_cols=["mean_companies","sd_companies","mean_others","sd_others",
                "companies_minus_others","se","t","cohens_d"],
    digits=2
)

# ------------------------------------------------------------
# FIELD-STANDARD TESTS â€” Paired t-tests per domain
# ------------------------------------------------------------
agg = (
    long.assign(group=np.where(long["actor"].isin(COMPANY_SET), "Companies", "Others"))
        .groupby(["resp_id","domain","group"], as_index=False, observed=False)
        .agg(m=("value","mean"))
)

tt_rows = []
for dom in DOMAINS:
    sub = agg[agg["domain"] == dom].pivot(index="resp_id", columns="group", values="m")
    sub = sub.dropna()  # respondents with both group means in this domain

    # Paired t-test
    tstat, pval = stats.ttest_rel(sub["Companies"], sub["Others"])

    # Mean difference & CI
    diff = sub["Companies"] - sub["Others"]
    n_pairs = diff.shape[0]
    mdiff = diff.mean()
    sd_diff = diff.std(ddof=1)
    se_diff = sd_diff / np.sqrt(n_pairs) if n_pairs > 0 else np.nan
    tcrit = stats.t.ppf(0.975, df=max(n_pairs-1,1))
    ci_low = mdiff - tcrit*se_diff
    ci_high = mdiff + tcrit*se_diff

    # Cohen's d_z for paired design
    dz = (mdiff / sd_diff) if sd_diff > 0 else np.nan

    # Group means/SDs (within-person means)
    m_comp, sd_comp = sub["Companies"].mean(), sub["Companies"].std(ddof=1)
    m_oth,  sd_oth  = sub["Others"].mean(),    sub["Others"].std(ddof=1)

    tt_rows.append({
        "domain": dom,
        "n_pairs": n_pairs,
        "M_companies": m_comp, "SD_companies": sd_comp,
        "M_others": m_oth, "SD_others": sd_oth,
        "mean_diff": mdiff, "SE_diff": se_diff,
        "CI_low": ci_low, "CI_high": ci_high,
        "t": tstat, "p": pval, "cohens_dz": dz
    })

tt = pd.DataFrame(tt_rows)

# Holm correction across the three paired t-tests
rej_tt, p_holm_tt, _, _ = multipletests(tt["p"].values, method="holm")
tt["p_holm"] = p_holm_tt
tt["sig_holm"] = rej_tt

# Display (rounded) â€” vectorized mapping for p-values
tt_disp = tt.assign(
    M_companies=lambda x: x["M_companies"].round(2),
    SD_companies=lambda x: x["SD_companies"].round(2),
    M_others=lambda x: x["M_others"].round(2),
    SD_others=lambda x: x["SD_others"].round(2),
    mean_diff=lambda x: x["mean_diff"].round(2),
    SE_diff=lambda x: x["SE_diff"].round(2),
    CI_low=lambda x: x["CI_low"].round(2),
    CI_high=lambda x: x["CI_high"].round(2),
    t=lambda x: x["t"].round(2),
    p=lambda x: x["p"].map(lambda v: "<.001" if v < .001 else f"{v:.3f}"),
    p_holm=lambda x: x["p_holm"].map(lambda v: "<.001" if v < .001 else f"{v:.3f}"),
    cohens_dz=lambda x: x["cohens_dz"].round(2),
)

display(Markdown("### Paired t-tests (Companies vs Others) within domain"))
show_table(tt_disp[[
    "domain","n_pairs",
    "M_companies","SD_companies","M_others","SD_others",
    "mean_diff","SE_diff","CI_low","CI_high","t","p","p_holm","cohens_dz"
]])

# Export CSV
rd_write_csv_df(tt, rd_join(OUT_DIR, "rq1_paired_ttests_companies_vs_others.csv"), index=False)

# APA-style lines (safe formatting)
print("\nAPA-style (paired t-tests):")
for _, r in tt.iterrows():
    df_t = int(r["n_pairs"] - 1) if pd.notna(r["n_pairs"]) else "df"
    p_str = "<.001" if r["p"] < .001 else f"{r['p']:.3f}"
    p_holm_str = "<.001" if r["p_holm"] < .001 else f"{r['p_holm']:.3f}"
    print(
        f"{r['domain']}: Companies (M={r['M_companies']:.2f}, SD={r['SD_companies']:.2f}) vs "
        f"Others (M={r['M_others']:.2f}, SD={r['SD_others']:.2f}); "
        f"Î”M={r['mean_diff']:.2f}, 95% CI [{r['CI_low']:.2f}, {r['CI_high']:.2f}], "
        f"t({df_t})={r['t']:.2f}, p={p_str} (Holm={p_holm_str}), d_z={r['cohens_dz']:.2f}."
    )

# ------------------------------------------------------------
# 2Ã—3 REPEATED-MEASURES ANOVA (Group Ã— Domain)
# ------------------------------------------------------------
# Keep only respondents with complete data (both groups in all three domains)
wide = agg.pivot_table(index="resp_id", columns=["domain","group"], values="m")
complete = wide.dropna()  # balanced set for AnovaRM
aov_long = (
    complete.stack(["domain","group"])
            .rename("m").reset_index()
)

aov = AnovaRM(
    data=aov_long, depvar="m", subject="resp_id",
    within=["group","domain"]
).fit()

print("\nRepeated-measures ANOVA (Group Ã— Domain):")
print(aov.summary())

# Partial eta-squared for each effect
def partial_eta_squared(F, df1, df2):
    try:
        return (F*df1) / (F*df1 + df2)
    except Exception:
        return np.nan

try:
    aov_tbl = aov.anova_table.reset_index().rename(
        columns={"index":"Effect","F Value":"F","Num DF":"df1","Den DF":"df2","Pr > F":"p"}
    )
    aov_tbl["pes"] = aov_tbl.apply(lambda r: partial_eta_squared(r["F"], r["df1"], r["df2"]), axis=1)
    aov_disp = aov_tbl.assign(
        F=lambda x: x["F"].round(2),
        df1=lambda x: x["df1"].astype(int),
        df2=lambda x: x["df2"].astype(int),
        p=lambda x: ["<.001" if v<.001 else f"{v:.3f}" for v in x["p"]],
        pes=lambda x: x["pes"].round(3)
    )[["Effect","df1","df2","F","p","pes"]]

    display(Markdown("### Repeated-measures ANOVA (summary with partial Î·Â²)"))
    show_table(aov_disp)

    # Export
    rd_write_csv_df(aov_tbl, rd_join(OUT_DIR, "rq1_rmANOVA_group_by_domain.csv"), index=False)

    # APA-style lines for ANOVA
    print("\nAPA-style (ANOVA key effects):")
    for _, r in aov_disp.iterrows():
        print(f"{r['Effect']}: F({r['df1']}, {r['df2']}) = {r['F']}, p = {r['p']}, partial Î·Â² = {r['pes']}.")
except Exception as e:
    print("ANOVA table parsing note:", e)

print(f"\nâœ… Done. Outputs written under remote OUT_DIR: {OUT_DIR}")
````

````python
from scipy import stats
from statsmodels.stats.multitest import multipletests
import pandas as pd
import numpy as np

def company_vs_actor_contrasts(long, domain, actor_labels=ACTOR_ORDER):
    sub = long[long["domain"]==domain].copy()
    sub["is_company"] = sub["actor"].isin([
        "Tech providers (developers)",
        "Tech users (adopting firms)"
    ]).astype(int)

    rows = []
    for target in actor_labels:
        if target in {"Tech providers (developers)","Tech users (adopting firms)"}:
            continue  # skip comparing companies to themselves

        # Keep only company rows + the target actor rows
        pair = sub[sub["actor"].isin({target, 
                                      "Tech providers (developers)",
                                      "Tech users (adopting firms)"})].copy()

        fit = smf.ols("value ~ is_company", data=pair).fit(
            cov_type="cluster", cov_kwds={"groups": pair["resp_id"]}
        )

        b  = fit.params.get("is_company", np.nan)
        se = fit.bse.get("is_company", np.nan)
        t  = b/se if (pd.notna(se) and se>0) else np.nan
        p  = 2*stats.t.sf(abs(t), df=fit.df_resid) if pd.notna(t) else np.nan

        # descriptive group means
        g1 = pair[pair["is_company"]==1]["value"]
        g0 = pair[pair["is_company"]==0]["value"]  # this is the single target actor
        m_comp = g1.mean(); m_targ = g0.mean()

        # Cohen's d (pooled SD)
        sp = np.sqrt(((len(g1)-1)*g1.var(ddof=1) + (len(g0)-1)*g0.var(ddof=1)) / max(len(g1)+len(g0)-2, 1))
        d_eff = (m_comp - m_targ)/sp if sp>0 else np.nan

        rows.append({
            "domain": domain,
            "comparison": f"Companies âˆ’ {target}",
            "b_diff": b, "se": se, "t": t, "p": p,
            "mean_companies": m_comp, "mean_target": m_targ,
            "cohens_d": d_eff,
            "n_companies": len(g1), "n_target": len(g0)
        })

    out = pd.DataFrame(rows)
    # Holm correction within domain (family = all pairwise vs actors in this domain)
    rej, p_holm, _, _ = multipletests(out["p"].astype(float).values, method="holm")
    out["p_holm"] = p_holm
    out["sig_holm"] = rej
    return out

# build and save the full table
contrasts_all = pd.concat([company_vs_actor_contrasts(long, d) 
                           for d in ["Environmental","Social","Governance"]],
                          ignore_index=True)

# nice rounding for display
tbl = contrasts_all.copy()
for c in ["b_diff","se","t","cohens_d","mean_companies","mean_target"]:
    tbl[c] = tbl[c].round(2)
tbl["p"] = tbl["p"].map(lambda v: "<.001" if pd.notna(v) and v<.001 else (f"{v:.3f}" if pd.notna(v) else "NA"))
tbl["p_holm"] = tbl["p_holm"].map(lambda v: "<.001" if pd.notna(v) and v<.001 else (f"{v:.3f}" if pd.notna(v) else "NA"))

rd_write_csv_df(contrasts_all, rd_join(OUT_DIR, "rq1_companies_vs_each_actor_by_domain.csv"), index=False)
display(Markdown("#### Companies vs each actor (cluster-robust OLS), Holm-corrected within domain"))
show_table(tbl[["domain","comparison","b_diff","se","t","p","p_holm","cohens_d","mean_companies","mean_target"]])
````
